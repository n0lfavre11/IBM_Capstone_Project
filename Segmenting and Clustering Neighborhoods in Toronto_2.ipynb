```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url,'lxml')
```

```python
table_post = soup.find('table')
fields = table_post.find_all('td')

postcode = []
borough = []
neighborhood = []

for i in range(0, len(fields), 3):
    postcode.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighborhood.append(fields[i+2].text.strip())
        
df_pc = pd.DataFrame(data=[postcode, borough, neighborhood]).transpose()
df_pc.columns = ['Postcode', 'Borough', 'Neighborhood']
df_pc.head()
```

```python
df_pc.head()
```

```python
df_pc['Borough'].replace('Not assigned', np.nan, inplace=True)
df_pc.dropna(subset=['Borough'], inplace=True)
df_pc.head()
```

```python
df_pcn = df_pc.groupby(['Postcode', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
df_pcn.columns = ['Postcode', 'Borough', 'Neighborhood']
df_pcn
```

```python
df_pcn['Neighborhood'].replace('Not assigned', "Queen's Park", inplace=True)
df_pcn
```

```python
# one hot encoding
df_tor_cen_oh = pd.get_dummies(df_ven[['VenueCategory']], prefix="", prefix_sep="")

# add postcode, borough and neighborhood column back to dataframe
df_tor_cen_oh['Postcode'] = df_ven['Postcode'] 
df_tor_cen_oh['Borough'] = df_ven['Borough'] 
df_tor_cen_oh['Neighborhoods'] = df_ven['Neighborhood'] 

# move postal, borough and neighborhood column to the first column
fixed_columns = list(df_tor_cen_oh.columns[-3:]) + list(df_tor_cen_oh.columns[:-3])
df_tor_cen_oh = df_tor_cen_oh[fixed_columns]

print(df_tor_cen_oh.shape)
df_tor_cen_oh.head()
```

```python
df_tor_cen_ven_f = df_tor_cen_oh.groupby(['Postcode', 'Borough', 'Neighborhoods']).mean().reset_index()
print(df_tor_cen_ven_f.shape)
df_tor_cen_ven_f.head()
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
areaColumns = ['Postcode', 'Borough', 'Neighborhoods']
freqColumns = []
for ind in np.arange(num_top_venues):
    try:
        freqColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freqColumns.append('{}th Most Common Venue'.format(ind+1))
        
columns = areaColumns + freqColumns
# create a new dataframe
n_ven_sorted = pd.DataFrame(columns=columns)
n_ven_sorted['Postcode'] = df_tor_cen_ven_f['Postcode']
n_ven_sorted['Borough'] = df_tor_cen_ven_f['Borough']
n_ven_sorted['Neighborhoods'] = df_tor_cen_ven_f['Neighborhoods']

for ind in np.arange(df_tor_cen_ven_f.shape[0]):
    r_cat = df_tor_cen_ven_f.iloc[ind, :].iloc[3:]
    r_cat_sorted = r_cat.sort_values(ascending=False)
    n_ven_sorted.iloc[ind, 3:] = r_cat_sorted.index.values[0:num_top_venues]

n_ven_sorted.sort_values(freqColumns, inplace=True)
n_ven_sorted
```

```python
k = 3

df_tor_cen_ven_f_cluster = df_tor_cen_ven_f.drop(['Postcode', 'Borough', 'Neighborhoods'], 1)

kmeans = KMeans(n_clusters=k, random_state=0).fit(df_tor_cen_ven_f_cluster)
df_tor_cen_cluster = df_tor_cen
df_tor_cen_cluster['Cluster'] = kmeans.labels_

df_tor_cen_cluster = df_tor_cen_cluster.join(n_ven_sorted.drop(['Borough', 'Neighborhoods'], 1).set_index('Postcode'), on='Postcode')
df_tor_cen_cluster.sort_values(['Cluster'] + freqColumns, inplace=True)
df_tor_cen_cluster
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(k)
ys = [i+x+(i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, post, bor, poi, cluster in zip(df_tor_cen_cluster['Latitude'], df_tor_cen_cluster['Longitude'], df_tor_cen_cluster['Postcode'], df_tor_cen_cluster['Borough'], df_tor_cen_cluster['Neighborhood'], df_tor_cen_cluster['Cluster']):
    label = folium.Popup('{} ({}): {} - Cluster {}'.format(bor, post, poi, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
